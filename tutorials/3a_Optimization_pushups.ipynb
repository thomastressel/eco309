__Write a function `bisection(f::Function, a::Float64, b::Float64)` which computes a zero of function `f` within `(a,b)` using a bisection method.__

```julia
function bisection(f::Function, a::Float64, b::Float64; ε=1e-8, δ=1e-8)
    if f(a) == 0
        return a
    elseif f(b) == 0
        return b
    elseif f(a)*f(b) > 0
        error("No solution")
    end
    n = 0
    a_n = a
    b_n = b
    c_n = (a + b) / 2
    fc = f(c_n)
    while abs(fc) > ε && (b - a)/2^n > δ
        c_n = (a_n + b_n) / 2
        fc = f(c_n)
        if fc * f(a_n) < 0
            b_n = c_n
        else
            a_n = c_n
        end
        n += 1
    end
    return c_n
end
```

```julia
function f(q)
    return exp(-q)*q - (exp(-q)*q - exp(-q))*q + (2*log(2) - log(q)) - 2
end

zero = bisection(f, 0.4, 1.0)
```

```julia
pl = plot()
plot!(pl, f, 0.5, 1.0, label = "")
vline!([zero], label = "Solution")
```
